```python
import os

# Base directory where the folders will be created
base_dir = "Data\\raster\\addiation_process_1990\\monthly"

# Ensure the base directory exists
os.makedirs(base_dir, exist_ok=True)

# List of region names
region_names = [
    "qld_et_short",
    "qld_max_temp",
    "qld_min_temp",
    "qld_rainfall",
    "qld_rh_tmax",
    "qld_rh_tmin",
    "qld_solar_radiation",
    "qld_vapour_pressure",
    "qld_vp_deficit"
]

# Create directories for each region
for region in region_names:
    region_dir = os.path.join(base_dir, region.replace(" ", "_"))
    os.makedirs(region_dir, exist_ok=True)
```
